```python
# Example to implement a simple neuron that has 2 inputs and sigmoid activation function #
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
class Neuron:
    def __init__(self,weights, bias):
        self.weights = weights
        self.bias = bias
    
    def feedforward(self, x):
        intermediary_value = np.dot(self.weights, x) + self.bias
        return sigmoid(intermediary_value)   
```

```python
#Test 
weights = np.array([0, 1]) # w1 = 0, w2 = 1
bias = 4                   # b = 4
n = Neuron(weights, bias)

x = np.array([2, 3])       # x1 = 2, x2 = 3
print(n.feedforward(x))    # 0.9990889488055994
```
